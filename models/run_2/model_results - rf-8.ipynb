```python
#standard libraries
import pandas as pd, numpy as np

#visualization libraries
import seaborn as sns
import matplotlib as plt
from matplotlib import pyplot as plt
```

```python
model_results = pd.read_excel('pilots_random_forest_results_without_noise.xlsx', index_col=0)
model_results.head(3)
```

```python
#set the canvas
plt.figure(figsize=(15, 4))

#plot the lineplot
sns.lineplot(data=model_results,
             x='palm_file',
             y='F1-Score, test',
             markers=True,
             dashes=False,
             marker='o'
             ).set(title='F1-Score')

plt.xticks(rotation=90);
```

```python
#F1-Score
print(f"max: {model_results['F1-Score, test'].max()}, \n min: {model_results['F1-Score, test'].min()}")
```

```python
print(f"mean: {model_results['F1-Score, test'].mean().round()}, \n median: {model_results['F1-Score, test'].median()}")
```

```python
#set the canvas
plt.figure(figsize=(15, 4))

#plot the lineplot
sns.lineplot(data=model_results,
             x='palm_file',
             y='learning_time',
             markers=True,
             dashes=False,
             marker='o'
             ).set(title='learning time')

plt.xticks(rotation=90);
```

```python
#learning time
print(f"max: {model_results['learning_time'].max()}, \n min: {model_results['learning_time'].min()}")
```
